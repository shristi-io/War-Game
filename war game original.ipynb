```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
class Card():
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit

class Deck():
    def __init__(self):
        self.cards = []
        
        for suit in suits:
            for rank in ranks:
                self.cards.append(Card(suit,rank))
    
    def shuffle(self):
        random.shuffle(self.cards)
        
    def deal_card(self):
        return self.cards.pop()
    
class Player():
    
    def __init__(self, name):
        self.name = name
        self.cards = []
        
    def remove_card(self):
        return self.cards.pop(0)
    
    def add_card(self, cards):
        if type(cards)==type([]):
            self.cards.extend(cards)
        else:
            self.cards.append(cards)
            
        def __str__(self):
            return f"Player {self.name} has {len(self.cards)} cards"
        
    print("Let's start war")
deck = Deck()
deck.shuffle()

player1 = Player("Kuku")
player2 = Player("Kuki")

for card in range(26):
    player1.add_card(deck.deal_card())
    player2.add_card(deck.deal_card())
print(len(player1.cards))
print(len(player2.cards))

round_no = 0

game_on = True
while game_on:
    
    round_no +=1
    print(f"\nRound {round_no}\n")
    player1_cards = []
    player2_cards = []
    
    if len(player1.cards) == 0:
        print(f"{player2.name} wins")
        game_on = False 
        break
    elif len(player2.cards) == 0:
        print(f"{player1.name} wins")
        game_on = False 
        break
    
    player1_cards.append(player1.remove_card())
    player2_cards.append(player2.remove_card())
    
    print(f"This is player 1 = {player1_cards[-1].value}")
    print(f"This is player 2 = {player2_cards[-1].value}")
    
    war = True
    while war:
        
            if player1_cards[-1].value > player2_cards[-1].value:
                player1.add_card(player2_cards)
                player1.add_card(player1_cards)
                war = False
                
            
                
            elif player1_cards[-1].value < player2_cards[-1].value:
                player2.add_card(player2_cards)
                player2.add_card(player1_cards)
                war = False
                
            else:
                print("War begins")
                if len(player1.cards) < 10:
                    print(f"{player1.name} out of cards! {player2.name} wins!")
                    game_on = False
                    war = False
                    
                elif len(player2.cards) < 10:
                    print(f"{player2.name} out of cards! {player1.name} wins!")
                    game_on = False
                    war = False
                
                else:
                    for num in range(10):
                        player1_cards.append(player1.remove_card())
                        player1_cards.append(player2.remove_card())
```
